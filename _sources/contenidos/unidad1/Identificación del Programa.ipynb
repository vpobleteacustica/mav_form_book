<table align="left" style="width:95%">

  <tr style="outline: thin solid">
      <td style="width: 7px;background:#036ED1;"></td>
  <td bgcolor="white"><div style="text-align: justify; ">El Magíster en Acústica y Vibraciones (MAV) contempla desde el inicio de sus actividades la articulación pregrado-postgrado, específicamente, con la carrera de Ingeniería Civil Acústica de la Universidad Austral de Chile (Anexos 10 y 11).</div><br>

<div style="text-align: justify; ">La articulación se rige, en el nivel institucional, por los Lineamientos para la Vinculación Pregrado-Postgrado de 2010 (Anexo 3.4); las estipulaciones del Reglamento General de Magíster de 2014 (Anexo 3.1); y el Procedimiento Académico-Administrativo para Programas de Vinculación Pregrado-Magíster actualizado en 2017 (Anexo 3.6). En el nivel operativo, los términos de la articulación se formalizan a través de "Acuerdos de Vinculación". Estos "Acuerdos" establecen la duración del programa de articulación pregrado-postgrado (i.e., la fase pregrado de la articulación), los cupos disponibles, los requisitos de admisión y las asignaturas o módulos que serán convalidados.</div><br> 

<div style="text-align: justify; ">La articulación específica del MAV con la carrera de Ingeniería Civil Acústica se rigió en una primera etapa por el Programa de Vinculación de la Facultad de Ciencias de la Ingeniería vigente hasta 2013. Dicho Acuerdo se actualizó primero en 2014 (Anexo 10) y más recientemente en 2021 (Anexo 11).  La modalidad actualmente en vigencia en el programa se rige por este último e incluye como innovación la gestión de las vinculaciones a través de una plataforma en línea desarrollada en 2019 desde la Vicerrectoría Académica. Los aspectos específicos del mecanismo de articulación y sus resultados se detallan en las secciones 3.2.6 y 3.3.3.</div></td>
