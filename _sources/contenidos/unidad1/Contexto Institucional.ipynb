# Contexto Institucional
## Entorno institucional
### Pertinencia de la actividad en el contexto académico en el que actúa la universidad que ofrece el programa

<table align="left" style="width:100%">    
  <tbody style="font-size: 14px";>
  <tr style="outline: thin solid">
      <td style="width: 7px;background:#036ED1;"></td>
  <td bgcolor="white"><div style="text-align: justify; ">El Magíster en Acústica y Vibraciones (MAV) contempla desde el inicio de sus actividades la articulación pregrado-postgrado, específicamente, con la carrera de Ingeniería Civil Acústica de la Universidad Austral de Chile (Anexos 10 y 11).</div><br>

<div style="text-align: justify; ">La Universidad Austral de Chile, con una trayectoria de 68 años, posee todas las características de una Universidad compleja en las que las actividades de docencia, investigación y vinculación con el medio se entrelazan en su quehacer diario. La Universidad cuenta con una tradición de más de cuarenta años en la formación de postgrado y postítulo y una consolidada política de formación de recursos humanos de alto nivel.</div><br> 

<div style="text-align: justify; ">La universidad ha definido a la investigación científica y la formación de postgrado como pilares de su Plan Estratégico Institucional 2020-2023 (Anexo 12). Existe además una estrecha vinculación de los programas de postgrado con las principales áreas de desarrollo en investigación a nivel institucional, lo que resulta en un modelo que asegura la calidad del postgrado.</div><br> 

<div style="text-align: justify; ">En la actualidad, la universidad imparte, a través de sus campus y sedes en las ciudades de Valdivia,  Puerto Montt y Coyhaique, once programas de doctorado, treinta y cuatro programas de magíster, dieciséis especialidades médicas, cuatro especialidades en ciencias clínicas veterinarias y veinticinco diplomados. En todos los niveles de programas con tasas de acreditación sobresalientes. A diciembre de 2021, la tasa de acreditación era de 100% en el caso de doctorado y 76% en magíster. Esta oferta de docencia de postgrado y postítulo se desarrolla en un ambiente multidisciplinario, en un entorno moderno y armónico con las inquietudes y necesidades de los estudiantes.</div><br> 

<div style="text-align: justify; ">La institución está inserta en una región de gran riqueza natural con componentes propios de la zona sur, estableciendo un compromiso por el conocimiento, la naturaleza y el desarrollo sustentable como una competencia sello de los alumnos de la institución.</div><br>  

<div style="text-align: justify; ">Los orígenes de lo que actualmente es la Facultad de Ciencias de la Ingeniería (FCI) de la Universidad Austral de Chile se remontan a 1934, fecha en que se crea la Escuela Industrial de Valdivia que tenía por misión preparar a operarios especializados. En 1947, y ante el auge de la enseñanza técnica, pasó a ser la sede Valdivia de la Universidad Técnica del Estado. A partir de 1981, luego de la reforma universitaria, se convirtió en el Instituto Profesional de Valdivia (IPV) y, finalmente, el año 1989 el ex IPV se integra a la Universidad Austral de Chile dando vida a la Facultad de Ciencias de la Ingeniería.  Es en esa oportunidad que la carrera de Ingeniería Acústica y el Instituto de Acústica que pertenecían a la Facultad de Ciencias, se trasladan de Campus y pasan a formar parte de la nueva FCI.</div><br>

<div style="text-align: justify; ">Actualmente, la FCI cuenta con ocho carreras de pregrado: Ingeniería Civil Acústica, Ingeniería Civil Electrónica, Ingeniería en Construcción, Ingeniería Civil en Obras Civiles, Ingeniería Naval, Ingeniería Civil Industrial, Ingeniería Civil en Informática e Ingeniería Civil Mecánica y un Bachillerato en Ciencias de la Ingeniería, que tiene una duración de dos años.</div><br>  

<div style="text-align: justify; ">En el nivel de postgrado, la FCI imparte el Magíster en Acústica y Vibraciones, el Magíster en Ingeniería Mecánica y Materiales, el Magíster en Informática, el Magíster en Ingeniería Naval y Oceánica, el Magíster en Electricidad y Electrónica Aplicada y el Diplomado en Eficiencia Energética y Calidad Ambiental en la Construcción.</div><br>

<div style="text-align: justify; ">La pertinencia del programa se contextualiza a partir de los ejes de compromisos para postgrado declarados en el Plan Estratégico Institucional 2020-2023 (Anexo 12) y su concordancia con el Plan Estratégico de la FCI 2020-2023 (Anexo 13) que declara como propósito estratégico fortalecer la calidad de los programas y sus cuerpos docentes a través de la internacionalización.</div><br>

<div style="text-align: justify; ">El programa de Magíster en Acústica y Vibraciones está vinculado fuertemente a la actividad académica y de investigación del Instituto de Acústica, unidad que ha liderado el desarrollo profesional de esta ciencia en Chile desde sus orígenes a principios de los años setenta. En los últimos años, el Instituto de Acústica ha venido desarrollando un plan de perfeccionamiento de sus académicos, lo que ha permitido que casi toda su planta docente posea estudios de doctorado. De igual forma, ha mantenido un constante incremento en su producción científica, vinculación con el sector socioproductivo y una intensa red de colaboración de sus académicos con el quehacer internacional en el área de la acústica y las vibraciones.</div><br>

<div style="text-align: justify; ">El año 2007 surgió la idea de crear un programa de magíster en estas dos áreas que permitiría desarrollar y proyectar la docencia del pregrado, la formación avanzada de recursos humanos, y la investigación en temas emergentes dentro del ámbito de la vibroacústica. El diseño del programa incorporó interdisciplinaridad integrando académicos de otros institutos de la Facultad de Ciencias de la Ingeniería, de la Facultad de Ciencias y de la Facultad de Filosofía y Humanidades, quienes han desarrollado investigación, aplicaciones e innovaciones de alto nivel, relacionadas con la acústica, las vibraciones, la fonética, las termociencias y las ciencias de los materiales. Esto se enmarca en la visión del Plan Estratégico de la Facultad que establece: “como Facultad compleja, se considera el trabajo académico interdisciplinario como un compromiso intelectual y ético en la preservación, generación y difusión del conocimiento.” Además, las temáticas abordadas dentro del programa pertenecen a las áreas de desarrollo estratégico, definidas en el Plan (Anexo 13).</div><br>

<div style="text-align: justify; ">Una motivación relevante en el origen de este programa era la demanda de estudios de postgrado en Acústica y Vibraciones por parte de los alumnos más talentosos, egresados de la carrera de Ingeniería Acústica e Ingeniería Civil Acústica de la propia Universidad Austral de Chile. Entonces, existía dificultad de perfeccionamiento en el extranjero debido al alto costo de estos programas, el incremento en la competitividad por becas y en la dificultad de reconocimiento de los estudios equivalentes a las ingenierías civiles chilenas por parte de las universidades de mayor prestigio en el mundo. De esta forma, y considerando que la propia Facultad de Ciencias de la Ingeniería ha establecido la necesidad de crear alternativas para que los estudiantes de las carreras de pregrado pudiesen continuar su formación en programas de magíster, es que este programa ha posibilitado exitosamente la vinculación pregrado- postgrado en la disciplina.</div><br> 

<div style="text-align: justify; ">En el diseño del magíster, se tuvo a la vista la experiencia de la continuidad de estudios de postgrado de los ingenieros titulados de la carrera de Ingeniería Acústica y de la propia experiencia de los profesores del Instituto de Acústica en universidades extranjeras como, por ejemplo, la Univ. Federal de Santa Caterina, Brasil; Univ. Federal de Río de Janeiro, Brasil; Univ. of New York, Estados Unidos; Auburn University-Alabama, Estados Unidos; Univ. Técnica de Berlín, Alemania; Danish Technical University, Dinamarca; Aalborg University, Dinamarca; Univ. of Dublin, Reino Unido; Univ. de Southampton, Reino Unido; Univ. de York, Reino Unido; Univ. Politécnica de Madrid, España; Univ. Politécnica de Valencia, España; Chalmers Univ. of Technology, Suecia; y la Univ. of Canterbury, Nueva Zelanda.</div><br>

<div style="text-align: justify; ">Con este programa de magíster, la Universidad Austral de Chile se posicionó como líder en el campo de la acústica y vibraciones en el país, ya no sólo con su programa de pregrado, sino también en el nivel de postgrado. El resultado de los procesos de formación de especialistas ha dado paso a otro tipo de relación con otras universidades, donde varios de los graduados del magíster han continuado estudios de doctorado, o bien, en la integración de graduados del programa en empresas e instituciones públicas, realizando un aporte relevante al desarrollo de Chile desde la acústica y vibraciones.</div><br>

<div style="text-align: justify; ">Otro aspecto importante de acentuar del Magíster de Acústica y Vibraciones, es que la FCI se convirtió además en un referente de postgrado en el área, pues es el único en su tipo en Latinoamérica dictado en idioma español. Esta condición ha permitido atraer estudiantes extranjeros a Valdivia, ha facilitado la cooperación internacional y ha generado mejores posibilidades de perfeccionamiento a los alumnos egresados (al poseer el grado de Magíster, están en mejores condiciones de postular a becas para proseguir estudios doctorales en el extranjero). Destaca en este aspecto el acuerdo de continuación de estudios de doctorado en la Universidad Politécnica de Madrid (donde un estudiante graduado del programa ha continuado estudios de doctorado), o el proyecto de acuerdo de doble titulación (en trámite), con la Universidad Politécnica de Valencia.</div><br>

<div style="text-align: justify; ">Todo lo anterior satisface la visión expresada en el Plan Estratégico de la Facultad que establece: “Ser reconocidos como un referente en el país en el área de ingeniería, contribuyendo al desarrollo no sólo de una sociedad globalizada sino también multicultural y sustentable, mediante el saber científico-tecnológico y humanista” (Anexo 13).</div><br>

### Estructura organizativa y descripción de funciones de los integrantes del comité académico, u órgano equivalente, responsable de la gestión del programa 

<table align="left" style="width:100%">    
  <tbody style="font-size: 14px";>
  <tr style="outline: thin solid">
      <td style="width: 7px;background:#036ED1;"></td>
  <td bgcolor="white"><div style="text-align: justify; ">En la Universidad Austral de Chile, el sistema de organización interna de los programas de magíster se rige por las disposiciones establecidas en el Reglamento de Escuelas de Graduados promulgado en 2015 (Anexo 3.2).</div><br>
<div style="text-align: justify; "><strong><em>a) Estructura organizativa</em></strong></div><br>
<div style="text-align: justify; ">La gestión académica y supervisión específica del Magíster en Acústica y Vibraciones, en virtud de lo dispuesto en los Artículos 13-15 del Reglamento citado (Anexo 3.2), es responsabilidad del Director de Programa, que desempeña sus funciones asesorado por el Comité de Programa. Tanto el Director como los integrantes del Comité son nombrados por la Comisión Central de Magíster a propuesta justificada del Director de la Escuela de Graduados de la Facultad de Ciencias de la Ingeniería. La formalización del nombramiento se realiza mediante resolución de la Dirección de Estudios de Postgrado que es el órgano de la administración central del cual dependen las Escuelas de Graduados de cada Facultad o Sede y cuya función principal es planificar, dirigir y coordinar el quehacer de las Escuelas.</div><br>

<div style="text-align: justify; ">El Director de Programa debe pertenecer a la planta académica y estar acreditado por la Comisión Central de Magíster. Dura cuatro años en su cargo y puede ser propuesto para otros períodos. El Comité del Programa es el órgano académico responsable de asesorar al Director. Debe estar integrado por a lo menos tres profesores acreditados por la Comisión Central de Magíster y que representen las áreas disciplinarias del programa.</div><br>
<div style="text-align: justify; "><strong><em>b) Funciones y atribuciones de los integrantes</em></strong></div><br>      
  <ol>
  <li><div style="text-align: justify; ">Evaluar y supervisar la actividad del programa, además de realizar un informe anual a la Escuela de Graduados respecto de la acreditación del programa y sus indicadores académicos (graduación oportuna, matricula, permanencia, otros). </div></li>
  <li><div style="text-align: justify; ">Analizar y resolver respecto a los antecedentes de los postulantes al programa.</div></li>
  <li><div style="text-align: justify; ">Analizar y resolver en primera instancia las solicitudes de convalidación y reconocimiento de asignaturas.</div></li>
  <li><div style="text-align: justify; ">Proponer a la Escuela de Graduados modificaciones a los planes de estudios del programa. </div></li>
  <li><div style="text-align: justify; ">Formalizar las comisiones evaluadoras de tesis a proposición del profesor patrocinante respectivo.</div></li>
  <li><div style="text-align: justify; ">Establecer mecanismos de aseguramiento de la calidad del programa incluidos los procesos de acreditación ante instancias nacionales e internacionales, según corresponda.</div></li>
  <li><div style="text-align: justify; ">Toda otra función que le que le otorguen los Reglamentos.</div></li>
</ol><br>
<div style="text-align: justify; "><strong><em>c) Funcionamiento interno, normas y mecanismos</em></strong></div><br>
  <div style="text-align: justify; ">Para la ejecución de las funciones descritas en el apartado anterior, el Comité de Programa del Magíster en Acústica y Vibraciones realiza reuniones periódicas, no inferiores a dos por semestre, según la etapa de actividades del programa previstas en el Calendario Académico: evaluación de antecedentes de los postulantes, preparación de la oferta semestral de asignaturas, revisión del estado de la progresión académica de las cohortes en etapa lectiva, seguimiento de los avances de tesis, participación del claustro en actividades asociadas a las líneas de investigación del programa, análisis de situaciones puntuales relacionadas al plan de estudios, revisión de solicitudes de los estudiantes (convalidaciones de asignaturas, suspensiones de semestre, reincorporaciones).</div><br>

<div style="text-align: justify; ">La instancia colegiada dentro la Facultad de Ciencias de la Ingeniería de velar por el buen funcionamiento del Comité es el Consejo de Escuela de Graduados. Este Consejo lo integran el Director de la Escuela de Graduados, quien lo preside, los Directores de cada programa de postgrado de la Facultad y un representante de los estudiantes de postgrado elegido de entre de sus pares. El Consejo tiene por mandato realizar reuniones ordinarias a lo menos una vez cada tres meses y puede ser convocado en forma extraordinaria a iniciativa del Director de la Escuela o cuando lo solicite un tercio de sus integrantes.</div><br>
