### Objetivos del programa 

<table align="left" style="width:95%">    
  <tbody style="font-size: 14px";>
  <tr style="outline: thin solid">
      <td style="width: 7px;background:#036ED1;"></td>
  <td bgcolor="white"><div style="text-align: justify; ">El MAV tiene como principal objetivo formar postgraduados especialistas, que cuenten con conocimiento avanzado en el campo de acústica y vibraciones*, mediante la entrega de conceptos teóricos, actividades prácticas de laboratorios, estudio individual guiado y el desarrollo de una tesis de magíster. La tesis debe tener un carácter de originalidad en el tema o en la metodología y capacitará al graduado del programa para el perfeccionamiento, para la innovación tecnológica y para el desarrollo de investigación científica con un alto grado de independencia.</div><br>

<div style="text-align: justify; ">El programa entrega una formación apropiada tanto como para proseguir estudios doctorales como para proyectar y desarrollar actividades profesionales avanzadas en el desarrollo científico y tecnológico en acústica y vibraciones.</div><br>
<div style="text-align: justify; "><sup style="font-size: 14px;">*</sup>El alcance disciplinar del programa se establece en la <em>acústica</em> como ciencia que estudia el sonido y <em>vibraciones</em> como fenómeno físico vinculado con el sonido.</div>

### Áreas o Líneas de investigación o creación, según corresponda

<table align="left" style="width:95%">    
  <tbody style="font-size: 14px";>
  <tr style="outline: thin solid">
      <td style="width: 7px;background:#036ED1;"></td>
  <td bgcolor="white"><div style="text-align: justify; ">El programa contempla dos líneas de investigación: i) acústica y ii) vibraciones.</div><br>

  <strong><em>Acústica:</em></strong> Esta línea estudia modelar, analizar y predecir el comportamiento de sistemas acústicos de alta complejidad, resolver problemas prácticos proponiendo soluciones innovadoras y sustentables, a partir de un proceso de investigación científica, desarrollada de acuerdo a los procedimientos y exigencias que imponen las técnicas modernas en acústica. Se incluyen en esta área las sublíneas de instrumentación y sensores acústicos, acústica forense, análisis de señales acústicas, acústica ambiental, acústica musical, comportamiento de materiales acústicos, modelamiento acústico, y acústica de salas.

<strong><em>Vibraciones:</em></strong> Esta línea estudia modelar, analizar y predecir el comportamiento de sistemas vibratorios de alta complejidad, resolver problemas prácticos proponiendo soluciones innovadoras y sustentables, a partir de un proceso de investigación científica, desarrollada de acuerdo a los procedimientos y exigencias que imponen las técnicas modernas en vibraciones. Se incluyen en esta área las sublíneas de instrumentación y sensores en vibraciones, análisis de señales vibratorias, modelamiento numérico de vibraciones, comportamiento y propiedades vibratorias de materiales y diseño mecánico-vibratorio de materiales.

## Requisitos de admisión y proceso de selección
### Requisitos de admisión

<table align="left" style="width:100%">    
  <tbody style="font-size: 14px";>
  <tr style="outline: thin solid">
      <td style="width: 7px;background:#036ED1;"></td>
  <td bgcolor="white"><div style="text-align: justify; ">El Reglamento General de Programas de Magíster (Anexo 3.1: Art. 19) establece, como requisito transversal para el ingreso a programas de magíster en la universidad, estar en posesión del grado académico de licenciado o de un título profesional cuyo nivel y contenidos sean equivalentes a los requeridos para obtención de una licenciatura. El Reglamento estipula además que, en el caso de postulantes extranjeros [no hispanohablantes], es requisito acreditar un dominio adecuado de español.</div><br>

<div style="text-align: justify; ">Con base en esas estipulaciones generales, los interesados en ingresar al Magíster en Acústica y Vibraciones deben cumplir con los siguientes requisitos específicos de admisión:</div>     
  <ol>
  <li><div style="text-align: justify; ">Estar en posesión del grado académico de licenciado o de un título profesional en el área de las ciencias de la ingeniería cuyo nivel y contenidos sean equivalentes a los requeridos para la licenciatura. </div></li>
  <li><div style="text-align: justify; ">En casos excepcionales, el programa podría considerar la evaluación de postulantes con grados académicos o títulos profesionales en áreas afines a la acústica y vibraciones en disciplinas distintas de la ingeniería. En dichos casos, los postulantes deben acreditar una formación robusta en matemáticas, física y fundamentos de la acústica. La acreditación de esa formación previa puede realizarse mediante la certificación de las calificaciones en cursos aprobados sobre esas materias del modo descrito en el requisito 3 o a través de la aprobación satisfactoria de un plan de nivelación diseñado por el Comité de Programa a postulantes preseleccionados.</div></li>
  <li><div style="text-align: justify; ">Acreditar un rendimiento académico sobresaliente o destacado en los estudios realizados que refleje concordancia con las exigencias de conocimientos previos definidas en las actividades curriculares del Plan de Estudios del programa. Para el cumplimiento de este requisito, los postulantes deben presentar los certificados de notas de los cursos rendidos en los estudios previos, presentados como antecedentes en la postulación, con las correspondientes calificaciones e información sobre el sistema de calificación, visados por la institución de estudios que los emite.</div></li>
  <li><div style="text-align: justify; ">Presentar un Currículum Vitae que documente la trayectoria de actividades académicas y profesionales realizadas pertinentes y de relevancia al interés de profundización de estudios en el programa. </div></li>
  <li><div style="text-align: justify; ">Adjuntar dos cartas de recomendación o informes de académicos que avalen la postulación a un programa de este tipo.</div></li>
</ol><br>

<div style="text-align: justify; ">Los requisitos de admisión del MAV y el mecanismo de presentación de los documentos se informan: a) en el sitio web del programa <a href="#">www.magister.acusticauach.cl;</a> b) en los instrumentos de difusión de la Escuela de Graduados de la Facultad de Ciencias de la Ingeniería en sus distintos soportes (folletería impresa, insertos, posters, otros); y c) en la «Plataforma de Postulaciones en Línea» implementada por la Dirección de Estudios de Postgrado <a href="#">www.uach.cl/postgrado.</a></div><br>
    
<div style="text-align: justify; ">Las postulaciones se formalizan y activan mediante Solicitud de Admisión*, a través de la «Plataforma de Postulaciones en Línea». El proceso de postulación se completa una vez que la Solicitud de Admisión se <em>envía</em> a través de esta plataforma acompañada de todos los documentos especificados en los requisitos: a) los certificados de cada uno de los títulos y grados obtenidos y declarados en la postulación; b) las concentraciones de notas respectivas (en ambos casos, en documentos oficializados por la institución que otorgó el grado o título); d) el currículum vitae con los antecedentes relevantes de interés para el programa; y d) las cartas de recomendación o informes de académicos que avalan la postulación.</div><br>

<div style="text-align: justify; "><sup style="font-size: 14px;">*</sup>El periodo comprendido entre el inicio de actividades del programa en 2008 y el primer semestre de 2017, la Solicitud de Admisión y la presentación de los documentos se realizaba mediante su entrega directa al Comité de Programa, en formato papel (en persona o vía correo ordinario) o en fotocopia digital (vía correo electrónico). Con la habilitación de la «Plataforma de Postulaciones en Línea» a partir del segundo de 2017, el sistema de postulaciones registró un mejoramiento significativo, tanto por el incremento en el volumen de potenciales postulantes como por la calidad del sistema de registro y trazabilidad de la información, que facilita su seguimiento, y permite la construcción de indicadores para orientar la toma de decisiones. La plataforma desarrollada incluye, además del sistema de postulaciones, el espectro completo de actividades administrativas y curriculares de los programas de postgrado, desde la admisión hasta el egreso y graduación de los estudiantes.</div><br>

### Articulación pregrado-postgrado

<table align="left" style="width:100%">    
  <tbody style="font-size: 14px";>
  <tr style="outline: thin solid">
      <td style="width: 7px;background:#036ED1;"></td>
  <td bgcolor="white"><div style="text-align: justify; ">El Art. 48 del Reglamento General de Magíster (Anexo 3.1) dispone que los estudiantes de pregrado de la Universidad Austral de Chile pueden postular a programas de magíster a través de programas de vinculación entre su carrera de pregrado y el postgrado. La forma de ingreso y estructura del currículo se describe en las normativas desarrolladas para este efecto por la Vicerrectoría Académica de la universidad y por acuerdos específicos de vinculación entre una escuela de pregrado y el programa de magíster. Estos "Acuerdos" establecen la duración del programa de vinculación, los cupos disponibles, los requisitos de admisión y las asignaturas o módulos que serán convalidados. En la sección 1.3 se documentó que el Magíster en Acústica y Vibraciones contempla un sistema de vinculación pregrado-postgrado con la carrera de Ingeniería Civil Acústica de la Facultad de Ciencias de la Ingeniería. Este sistema se ha articulado, primero, con arreglo a los Lineamientos para la Vinculación Pregrado-Postgrado de 2010 (Anexo 3.4); al Procedimiento de Vinculación entre Carreras de Pregrado y Programas de Magíster de 2013 (Anexo 3.5); y a la actualización de ese Procedimiento en 2017 (Anexo 3.6).
</div><br>

<div style="text-align: justify; ">El programa contó con un primer Acuerdo en 2009 que estuvo vigente hasta 2013 y un segundo Acuerdo suscrito en 2014 que se mantuvo vigente hasta 2020 (Anexo 10). La modalidad actualmente en vigencia en el programa es el Acuerdo de Vinculación de 2021 (Anexo 11) que mantiene las principales características del Acuerdo de 2014 e incorpora como innovación la gestión y trazabilidad en línea de las vinculaciones. El Acuerdo establece los siguientes términos para los estudiantes interesados en ingresar al MAV a través de este sistema:</div><br>
    <ol type="a" >
      <li style="text-align: justify; ">Duración de la vinculación: dos semestres como máximo.</li>
      <li style="text-align: justify; ">Cupos anuales disponibles: ocho.</li>
      <li style="text-align: justify; ">Requisitos de ingreso: haber cumplido con los requisitos para la obtención de grado de Licenciado en Ciencias de la Ingeniería. </li>
      <li style="text-align: justify; ">Contar con un promedio general acumulado igual o superior a 5,0 (en una escala de 1 a 7) en la carrera de Ingeniería Civil Acústica. </li>
    </ol>
    <div style="text-align: justify; ">La verificación de los requisitos y selección de los postulantes la realiza el Comité de Programa en conjunto con el Director de la Escuela de Ingeniería Civil Acústica quien participa como invitado. Los detalles del sistema de postulación, selección, aprobación e incorporación final al programa como alumnos regulares de magíster se informan en los Anexos 3.6 y 11.</div><br>
      <div style="text-align: justify; ">Los estudiantes aceptados en la articulación pueden convalidar hasta un máximo de tres asignaturas de su plan de estudios de pregrado (detalladas en el Anexo 11) además de una actividad equivalente al trabajo terminal de su titulación en pregrado. A partir de 2018, es obligatorio cursar la actividad ACUS 397: Seminario de Tesis como trabajo final de titulación. Las asignaturas susceptibles de incluir en el programa pueden ser cualquiera de las disponibles en la malla curricular del Magíster en Acústica y Vibraciones. El estudiante define las asignaturas de su programa de articulación con el Director de Programa y con el Director de la Escuela de Ingeniería Civil Acústica.</div><br>
